### 2.Segregate the employees w.r.t different positions.

### 3.Find from which age group most of the employees belong to.

```python
labels=['20-29','30-39','40-49','50-59','60+']
df['Age_group']=pd.cut(df['Age'],bins=[20,30,40,50,60,100],labels=labels)


#COUNT EMPLOYEES IN EACH AGE GROUP
age_group_counts=df['Age_group'].value_counts()
max_age_group = df['Age_group'].value_counts().idxmax()
max_employees_count = df['Age_group'].value_counts().max()

print(f"The age group with the maximum number of employees is around {max_age_group} years with {max_employees_count} employees.")
```

### 4.Find out under which team and position, spending in terms of salary is high.

```python
# Group by 'Team' and 'Position' to sum up the salaries
team_position_salary = df.groupby(['Team', 'Position'])['Salary'].sum()

# Find the maximum spending and its corresponding team and position
max_spending = team_position_salary.max()
max_spending_info = team_position_salary.idxmax()

print(f"The highest spending in terms of salary is {max_spending} in the {max_spending_info[0]} team for the {max_spending_info[1]} position.")
```

### 5.Find if there is any correlation between age and salary , represent it visually.
